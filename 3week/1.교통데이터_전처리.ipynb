```python
# 승차하시각 타입변환 : 정수 --> 문자
df_bus_card_kor = df_bus_card_kor.astype({'승차시각':'str'})
df_bus_card_kor = df_bus_card_kor.astype({'하차시각':'str'})
df_bus_card_kor.info()
```

```python
# 분석에 필요한 항목을 타입변환하여 데이터프레임 추가 : 문자 --> 날짜
df_bus_cart = pd.DataFrame()

df_bus_cart['승차시각'] = pd.to_datetime(df_bus_card_kor['승차시각'])
df_bus_cart['하차시각'] = pd.to_datetime(df_bus_card_kor['하차시각'])
df_bus_cart['승객구분'] = df_bus_card_kor['승객연령']
df_bus_cart['환승여부'] = df_bus_card_kor['환승여부']
df_bus_cart['추가운임여부'] = df_bus_card_kor['추가운임여부']
df_bus_cart['승차정류장'] = df_bus_card_kor['승차정류장']
df_bus_cart['하차정류장'] = df_bus_card_kor['하차정류장']

df_bus_cart.head(3)
```

```python
# 체류시간
df_bus_cart.info()
```

```python
# 분단위로 데이터타입변환(분단위 - 소수 - 정수)
df_bus_cart['버스체류시간_분'] = (df_bus_cart['하차시각'] - df_bus_cart['승차시각']).astype('timedelta64[m]').astype('int')
df_bus_cart.info()
```

```python
# 1.데이터 통합하기

import os
file_path = './bus_crawling/01_data_원본/'

print('Start Time :',datetime.today().strftime('%y-%m-%d %H:%M:%S'))

for i in range(len(os.listdir(file_path))):
    f_path = './bus_crawling/01_data_원본/trfcard ('+ str(i) +')/trfcard.csv'
    csv_tmp = pd.read_csv(f_path)
    df_bus_card_tot = pd.concat([df_bus_card_tot,csv_tmp], ignore_index = True)

print('End Time :',datetime.today().strftime('%y-%m-%d %H:%M:%S'))    
```

```python
df_bus_card_tot.info()
```

```python
# 2.메타정보 변경
meta_path = './bus_crawling/01_data_원본/trfcard (0)/'
df_bus_card_col_org = pd.read_excel(meta_path + 'trfcard_columns.xlsx',header = 2, usecols=('B:C'))

bus_meta = {}

for x,v in zip(df_bus_card_col_org['컬럼명 (영문)'],df_bus_card_col_org['컬럼명 (한글)']):
    bus_meta[x] = v

df_bus_card_tot.rename(columns = bus_meta, inplace=True)
df_bus_card_tot.info()
```

```python
# 3.타입변경(숫자 - 문자 - 날짜)
print('Start Time :',datetime.today().strftime('%y-%m-%d %H:%M:%S'))

df_bus_card_tot['승차시각'] = pd.to_datetime(df_bus_card_tot['승차시각'].astype('string'))
df_bus_card_tot['하차시각'] = pd.to_datetime(df_bus_card_tot['하차시각'].astype('string'))

print('End Time :',datetime.today().strftime('%y-%m-%d %H:%M:%S'))

df_bus_card_tot.info()
```
