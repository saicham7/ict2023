```python
# 기준월 및 기준일별로 버스 이용량 확인
# 인덱스 : 기준월
# 컬럼 : 기준일
# 값 : 승객구분
# 함수 : count
df_pivot_d = df_bus_cart_tot.pivot_table(index ='기준월', columns='기준일', values = '승객구분', aggfunc='count'  )

# 결측값 0으로 정제
df_pivot_d = df_pivot_d.fillna(0)
df_pivot_d
```

```python
# 시각적으로 이벤트시점을 확인 : 히트맵(자료구조가 집계표여야 함)
plt.figure(figsize=(22,2))
sns.heatmap(df_pivot_d, cmap=plt.cm.Blues, annot = True, fmt ='.0f')
plt.title('기준월 및 일자별 버스 이용량 분석')
plt.show()
```

```python
# 기준일 및 시간별 버스이용량 분석
df_pivot_h = df_bus_cart_tot.pivot_table(index ='기준일', columns='기준시간', values = '승객구분', aggfunc='count'  )
df_pivot_h = df_pivot_h.fillna(0)

# 시각화
plt.figure(figsize=(22,11))
sns.heatmap(df_pivot_h, annot=True, fmt='.0f', cmap=plt.cm.Reds)
plt.title('기준일 및 시간별 버스이용량 분석')
plt.show()
```

```python
# 시간대별 승객구분별 버스내 체류시간 분석
df_temp = pd.DataFrame()
df_temp = df_bus_cart_tot[['기준시간','승객구분','버스체류시간_분']]
df_temp.head(3)
```

```python
# 시간대별 승객구분별 버스체류시간 분석
df_temp2 = (df_temp.groupby(['기준시간','승객구분'], as_index =False).sum()).sort_values(by='버스체류시간_분', ascending=False)
```

```python
df_temp2.head(10)
```

```python
# barplot : 범주형 데이터 확인시 용이하게 사용됨
# hue : 데이터 보여줄 값 설정(범주형)

plt.figure(figsize=(25,10))
sns.barplot(data = df_temp2, x='기준시간', y ='버스체류시간_분', hue ='승객구분')
plt.title("시간대별 승객구분별 버스내 체류시간 분석")
plt.show()
```

```python
import numpy as np
# 시간 및 승객구분별 빈도분석
plt.figure(figsize=(12,4))
g = sns.histplot(data=df_temp2, x='기준시간',bins=30, kde=True, hue ='승객구분', multiple = 'stack', stat='density', shrink = 0.6) #multiple = 'dodge','stack'
plt.title('시간 및 승객구분별 빈도분석')
#plt.xticks(np.arange(24))
plt.xlabel('기준시간')
plt.ylabel('rate(비율)')
plt.show()
```
