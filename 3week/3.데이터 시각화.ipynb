#### heatmap으로 시각화

```python
# 히트맵으로 년/월 데이터 상세 시각화
# 처리순서 : 피봇테이블 > 해시맵 시각화
# df.pivot_table(index =['년도'])
```

```python
# 중국  데이터를 이용해서 해시맵 시각화
df_filter = df[df['국적'] == '중국']
```

```python
df_pivot = df_filter.pivot_table(index=['년도'], # 행
                      columns = ['월'], # 열
                      values =['관광'], # 값
                      aggfunc='sum') # 집계함수
```

```python
# 해시맵 시각화를 위한 시각화 라이브러리 가져오기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 그래프 사이즈 설정
plt.figure(figsize=(16,10))

# 히트맵 그리기 : 피봇형태의 데이터를 넣어주어야 함
sns.heatmap(df_pivot, annot=True, fmt ='.0f', cmap = 'rocket_r')
plt.title('년월별 중국 관광객 상세')
plt.show()
```

```python
## 상위 5개 국가 추출
temp = df.pivot_table(index ='국적', values ='전체비율(%)', aggfunc='mean').sort_values(by='전체비율(%)', ascending=False).head(5)
temp.reset_index(inplace=True)

for natn in temp['국적']:
    # 그래프 화면 설정
    plt.figure(figsize=(16,10))
    # 그래프 함수 호출
    hamp_g(natn) 
    plt.show()    
```

```python
def hamp_g(natn):
    # 국가별 내역 추출
    df_filter = df[df['국적'] == natn]
    # 국가별 요약 집계
    df_hmap = df_filter.pivot_table(index = '년도',columns ='월', values ='관광', aggfunc='sum')
    # 히트맵 시각화
    sns.heatmap(df_hmap, annot = True,  fmt ='.0f', cmap = 'rocket_r')
    plt.title('{} 관광객 상세'.format(natn))
```
