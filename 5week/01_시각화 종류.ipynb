```python
import numpy as np
```

```python
q1, q3 = np.percentile(tip['total_bill'],[25,75])
```

```python
iqr = q3 - q1
mini = q1 - 1.5 * iqr 
maxi = q3 + 1.5 * iqr
iqr, mini, maxi
```

```python
df_l = list(tip[(tip['total_bill'] < mini) | (tip['total_bill'] > maxi)].index)
```

```python
tip[(tip['total_bill'] < mini) | (tip['total_bill'] > maxi) == True]
```

```python
for i in df_l:
    print(tip.loc[i,'total_bill'])
```

```python
sns.boxplot(tip['total_bill'])
```

```python
# 이상치 데이터 확인하기

def get_outlier(cols):
    q1, q3 = np.percentile(tip[cols],[25,75])
    iqr = q3 - q1
    mini = q1 - 1.5 * iqr 
    maxi = q3 + 1.5 * iqr
    rslt = tip[(tip[cols] < mini) | (tip[cols] > maxi) == True]
    return len(rslt), mini, maxi, cols, rslt
```

```python
cnt, mini, maxi, cols, rslt = get_outlier('total_bill')
print('항목 : {}  개수 : {}  최소값 : {} 최대값 : {}'.format(cols, cnt, mini, maxi))
print()
print(rslt)
```

```python
cnt, mini, maxi, cols, rslt = get_outlier('tip')
print('항목 : {}  개수 : {}  최소값 : {} 최대값 : {}'.format(cols, cnt, mini, maxi))
print()
print(rslt)
```

```python
cnt, mini, maxi, cols, rslt = get_outlier('size')
print('항목 : {}  개수 : {}  최소값 : {} 최대값 : {}'.format(cols, cnt, mini, maxi))
print()
print(rslt)
```
