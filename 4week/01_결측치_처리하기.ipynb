```python
import pandas as pd
df = pd.read_csv('./4day_전처리/data/bicycle.csv', encoding='euc-kr')
df
```

```python
df.info()
```

```python
# 결측치 확인하기
# isnull : True이면 결측데이터(null)이 있다는 의미
df['대여거치대'].isnull().sum()
```

```python
df.isnull() =='False'
#df.notnull()
```

```python
# 컬럼별 결측치 건수
df.isnull().sum(0)  # sum(0) 컬럼기준으로 합산, sum(1)  행기준으로 합산
```

```python
# 결측값 대체
# - 행삭제, 열삭제
# - 열 평균값, 주변데이터 입력, 임의값
```

```python
# 결측치 행 삭제
# dropna : axis 0 행삭제(default), axis 1 열삭제

# 결측값이 존재하는 행삭제
df.dropna(axis=0)
```

```python
# 결측값이 존재하는 컬럼삭제
df.dropna(axis=1)
```

```python
# 특정 컬럼의 결측치 삭제
df['대여거치대'].dropna(axis = 0)
```

```python
# 값이 존재하는 행건수
df['대여거치대'].notnull().sum()
```

```python
# 여러 컬럼을 기준으로 결측치 삭제
df[['대여거치대','대여소번호','이용시간']].dropna(axis = 0)
```

```python
# 결측치를 0으로 대체(문자, 숫자, 날자 항목 모두)
df.fillna(0)
```

```python
# 이용거리 결측값 0으로 대체
df['이용거리'].fillna(0)
```

```python
# 이용거리에 결측치가 있으면 모두 문자열(missing)로 대체하면 컬럼타입이 문자열로 변환됨
df['이용거리'].fillna('missing')
```

```python
# 결측값을 평균으로 대체
df.fillna(df.mean())
```

```python
# 이용거리 항목 결측값에 이용거리 평균값으로 대체
df['이용거리'].fillna(round(df['이용거리'].mean(),1))
```
