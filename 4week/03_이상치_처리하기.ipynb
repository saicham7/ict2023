### Outlier 확인하기

```python
# 25%, 75% 시점의 데이터 확인하기
# q1 = 25%
# q3 = 75%
import numpy as np
q1, q3 = np.percentile(df_drop_allrow['나이'],[25,75])
q1,q3
```

```python
# iqr, 최대값, 최소값 구하기
iqr = q3 - q1
upper_bound = q3 + (1.5 * iqr)
lower_bound = q1 - (1.5 * iqr)
iqr, upper_bound, lower_bound
```

```python
# 이상치 데이터 확인
df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound)]
```

```python
# 인덱스를(True/False) 이용해 이상치 데이터 필터링
len(df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound) == False])
```

```python
len(df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound) == True])
```

```python
# 이상치 데이터의 인덱스를 리스트로 뽑기
[df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound)].index]
```

```python
# 또는 list()를 입혀서 뽑기
list(df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound)].index)
```

```python
# numpy로 인덱스 확인
np.where((df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound))
```

```python
df_list = list(df_drop_allrow[(df_drop_allrow['나이'] > upper_bound) | (df_drop_allrow['나이'] < lower_bound)].index)
df_list
```

#### 데이터프레임.iloc[행,열]
#### 데이터프레임.loc[행,열명]

```python
# 데이터프레임.iloc[행,열]
# 데이터프레임.loc[행,열명]

for i in df_list:
    print(df_drop_allrow.iloc[i,11])
```

```python
# df_drop_allrow의 2번째 항목 추출
dd = df_drop_allrow.index
for i in dd:
    print(df_drop_allrow.loc[i,'반납일시'])
```

```python
# 평균값을 이상치 대체
age_av = int(df_drop_allrow['나이'].mean())

for i in df_list:
    df_drop_allrow.iloc[i,11] = age_av
```
