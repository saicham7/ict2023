## 특정 폴더의 전체 파일을 전처리후 통합하기

```python
import os
```

#### 1.특정 디렉토리의 특정 파일유형을 리스트에 넣기

```python
# 전처리 함수만들기
def tour_process(imsi_files):
    
    # 1. 대륙 제거
    natn_list = ['아시아주','미주','구주','대양주','아프리카주','기타대륙','교포소계']
    natn_cond = imsi_files['국적'].isin(natn_list)
    temp_files = imsi_files[natn_cond == False]
    
    # 2.index reset
    temp_files.reset_index(drop=True, inplace = True)
    
    # 3.대륙항목 생성
    continents = ['아시아'] * 25 + ['아메리카'] * 5 + ['유럽'] * 23 + \
                 ['오세아니아'] * 3 + ['아프리카'] * 2 + ['기타대륙','교포']
    temp_files['대륙'] = continents
    
    # 4.국가별 관광비율 생성
    temp_files['관광객비율(%)'] = round(temp_files['관광'] / temp_files['계'] * 100, 1)
    
    # 5.년도별 전체비율
    tour_sum = temp_files['관광'].sum()
    temp_files['전체비율(%)'] = round(temp_files['관광'] / tour_sum * 100, 1)
    
    # 6.값 반환
    return temp_files
```

#### 2.데이터 통합하기

```python
# 파일 합치기
total_kto = pd.DataFrame()

# 특정 경로의 파일 찾기
file_path = './02_데이터전처리/files/'
tot_file = os.listdir(file_path)

# 특정 파일유형만 찾기
xfile = [files for files in tot_file if files.endswith(".xlsx")]
#len(xfile)

# 데이터 통합하기
for files in xfile:
    
    # 1.파일읽기
    imsi_files = pd.read_excel(file_path + files, \
                               header=1,\
                               skipfooter = 4,\
                               usecols ="A:G")
    # 2.기준년월 생성
    imsi_files['기준년월'] = files[4:8] +'-'+ files[8:10]
                   
    # 3.전처리 함수 호출
    try:
        tour_files = tour_process(imsi_files) 
    # 4.데이터 합치기
        total_kto = pd.concat([total_kto,tour_files],ignore_index=True)
    except:
        pass
    
```
